```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
!wget https://lazyprogrammer.me/course_files/spam.csv
```

```python
df = pd.read_csv('spam.csv', encoding='ISO-8859-1')
```

```python
df.head()
```

```python
df = df.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
df.columns = ['labels', 'data']
```

```python
df.head()
```

```python
df['b_labels'] = df['labels'].map({'ham': 0, 'spam': 1})
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df['data'], df['b_labels'], test_size=0.2)
```

```python
X_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(decode_error='ignore')
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

#Logistic regression classifier

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score
train_pred = classifier.predict(X_train)
test_pred = classifier.predict(X_test)
```

```python
print('training set accuracy:', accuracy_score(y_train, train_pred))
```

```python
print('test set accuracy:', accuracy_score(y_test, test_pred))
```

#Decision tree classifier

```python
from sklearn.tree import DecisionTreeClassifier
dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
```

```python
train_pred = dt_classifier.predict(X_train)
test_pred = dt_classifier.predict(X_test)
print('training set accuracy:', accuracy_score(y_train, train_pred))
print('test set accuracy:', accuracy_score(y_test, test_pred))
```
